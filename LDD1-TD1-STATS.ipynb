### Dessinons la distribution théorique du t de Student utilisée dans le calcul de l'intervalle de confiance à 95% sur les données de Meng et al 2019

```r
# Calculons les degrés de liberté (ddl) de Meng et al 2019 à partir des données RMS
ddl <- length(RMS_data$Subject)-1
ddl
```

```r
#calculons la valeur unilatérale à 2.5% 
#(borne inférieure de l'intervalle de confiance à 95% pour la distribution théorique)
qt(0.025, ddl, lower.tail = TRUE, log.p = FALSE)
```

```r
plot(function(x) dt(x, df = ddl), -5, 5, ylim = c(0, 0.5),
     main = "Distribution du t(20) de Student", yaxs = "i")
```

### Ajoutons les bornes inférieures et supérieures de l'IC à 95%

```r
library(stats)

garantie=0.95
alpha_unilat=1-garantie
alpha_bilat=(1-garantie)/2


# ceci calcule la valeur du t de Student pour la borne inférieure de l'intervalle
x_lower <- qt(alpha_bilat, ddl, lower.tail = TRUE, log.p = FALSE)
x_lower <- round(x_lower, digits = 4)
x_lower

# ceci calcule la valeur de probabilité de densité pour la borne inférieure de l'intervalle
y_lower <- dt(x_lower, ddl)
y_lower


# ceci calcule la valeur du t de Student pour la borne supérieure de l'intervalle
x_upper <- qt(1-alpha_bilat, ddl, lower.tail = TRUE, log.p = FALSE)
x_upper <- round(x_upper, digits = 4)
x_upper

# ceci calcule la valeur de probabilité de densité pour la borne supérieure de l'intervalle
y_upper <- dt(x_upper, ddl)
y_upper
```

```r
plot(function(x) dt(x, df = ddl), -5, 5, ylim = c(0, 0.5),
     main = "Distribution du t(20) de Student", yaxs = "i")
abline(a=NULL, b=NULL, v=x_upper, col="blue", lwd=3, lty=3)
text(-3, 0.03, x_lower, cex = 2, col = "red")
abline(a=NULL, b=NULL,  v=x_lower, col="red", lwd=3, lty=3)
text(3, 0.03, x_upper, cex = 2, col = "blue")
```

```r
#install.packages("ggdist")
library(ggplot2)
library(dplyr)
library(ggdist)
```

```r
# degrés de liberté por la distribution du t de Student
ddl=20

# Make plots wider 
options(repr.plot.width=10, repr.plot.height=8)

My_Theme = theme(
  axis.title.x = element_text(size = 16),
  axis.text.x = element_text(size = 14),
  axis.text.y = element_text(size = 14),
  axis.title.y = element_text(size = 16),
   legend.title=element_text(size=16), 
    legend.text=element_text(size=14),
#plot.margin = margin(2,.8,2,.8, "cm")
)


tribble(
  ~ dist,      ~ args,
  "norm",      list(0, 1),
"student_t", list(ddl, 0, 1)
) %>%
  ggplot(aes(y = dist, dist = dist, args = args)) +
  stat_dist_halfeye() +
  ggtitle(paste("Distribution du t de Student (en haut) pour ", ddl, " degrés de liberté, ainsi que celle de la loi normale (en bas)", sep = "")) +
   My_Theme

```

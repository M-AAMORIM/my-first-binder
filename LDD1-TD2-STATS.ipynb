## TD2 LDD1 STATS du 9 février 2021

```r
RMS_data <- read.csv("Meng_et_al2019-RMS_data_TD1.csv", sep=";")
# on supprime les valeurs manquantes NA du tableau, avec la ligne suivante
RMS_data <- na.omit(RMS_data)
RMS_data
```

### On s'intéresse d'abord à l'effet *Vision* pour les RMS dans l'axe médiolatéral (x)
#### On extrait du tableau les données des variables correspondantes


```r
RMSdata_x <- RMS_data[grepl("x", names(RMS_data)) & !grepl("RQ", names(RMS_data))]

RMSdata_x
```

### on ajoute une colonne pour la moyenne des RMS_x *yeux fermés* et une autre pour *yeux ouverts*

```r
library(dplyr)

# on rajoute une colonne qui contient la moyenne par ligne des RMS avec rowMeans()
# des 3 colonnes de la condition "_novis" = closed eyes, avec select()
RMSdata_x$novis_RMS_x <- round(rowMeans(select(RMSdata_x, contains("_novis"))), digits=3)

# idem pour "_vis" = opened eyes
RMSdata_x$vis_RMS_x <- round(rowMeans(select(RMSdata_x, contains("_vis"))), digits=3)

RMSdata_x

```

### on rajoute une colonne pour l'effet *Vision* sur RMS_x

```r
RMSdata_x$VisEffect_RMS_x <- (RMSdata_x$novis_RMS_x - RMSdata_x$vis_RMS_x)

RMSdata_x
```

### On refait un tableau juste avec les nouvelles variables d'intérêt

```r
vis_effect_RMS_x <- data.frame(
  RMSdata_x$novis_RMS_x,
 RMSdata_x$vis_RMS_x,
  RMSdata_x$VisEffect_RMS_x
)
 
vis_effect_RMS_x
```

### On fait des stats descriptives rapides sur l'effet de la Vision sur les RMS en x

```r
library(psych)
resVisEffect_x <- round(describe(vis_effect_RMS_x, fast=TRUE, IQR=TRUE), digits=3)

resVisEffect_x
```

```r
# Load ggplot2
library(ggplot2)

# create dummy data
VisEffect_xdata <- data.frame(
  name=row.names(resVisEffect_x),
  value=resVisEffect_x$mean,
  se=resVisEffect_x$se
)
 
My_Theme = theme(
  axis.title.x = element_text(size = 16),
  axis.text.x = element_text(size = 14),
  axis.text.y = element_text(size = 14),
  axis.title.y = element_text(size = 16),
   legend.title=element_text(size=16), 
    legend.text=element_text(size=14),
)

# Most basic error bar
ggplot(VisEffect_xdata) + My_Theme +
    geom_bar( aes(x=name, y=value), stat="identity", fill="skyblue", alpha=0.7) +
labs(title = "Vision effect on MedioLateral RMS", x = NULL, y = "RMS (cm)") +
    geom_errorbar( aes(x=name, ymin=value-se, ymax=value+se), width=0.4, colour="orange", 
                  alpha=0.9, size=1.3) + 
scale_x_discrete(labels=c("Closed eyes","Opened eyes","Vision effect")) 
```

### Testons si l'effet *Vision* sur RMS_x est significatif 

#### On utilise un test *t* apparié, car il s'agit d'une mesure répétée (chaque participant est passé par les deux conditions *closed eyes* et *opened eyes*). L'effet *Vision* est donc un effet intra-sujet.

```r
# on utilise le data.frame *vis_effect_RMS_x* qui contient les données de chaque sujet

RMS_novis_x <- vis_effect_RMS_x$RMSdata_x.novis_RMS_x
RMS_vis_x <- vis_effect_RMS_x$RMSdata_x.vis_RMS_x

# on choisit un test t apparié (paired=TRUE) et bilatéral (alternative="two.sided")
VisEffect_t <- t.test(RMS_novis_x, RMS_vis_x, alternative="two.sided", paired=TRUE)
#alternative = “less” pour faire un test-t unilatéral inférieur (différence négative)
#alternative = “greater” pour faire un test-t unilatéral supérieur (différence positive)

summary(VisEffect_t)

effect_tvalue <- VisEffect_t$statistic # la valeur t calculée correspondant à l'effet étudié
effect_ddl <- VisEffect_t$parameter # degrés de liberté
effect_pvalue <- VisEffect_t$p.value #  valeur p du test de Student

effectCI <-  VisEffect_t$conf.int # intervalle de confiance (IC) autour de l'effet
effectCI_inf <- VisEffect_t$conf.int[1] # Valeur de la borne inférieure de l'IC autour de l'effet
effectCI_sup <- VisEffect_t$conf.int[2] # Valeur de la borne supérieure etc.

if (effect_pvalue < 0.05){
    cat("L'effet étudié est significatif au seuil alpha = 0.05, ")
} else {
    cat("L'effet étudié est non-significatif au seuil alpha = 0.05, ")
}

cat("t(", effect_ddl, ") = ", round(effect_tvalue, digits=4), ", p = ", +
    round(effect_pvalue, digits=5), ";", "\n")
    
   cat("l'intervalle de confiance (IC) à 95% autour de l'effet est [", +
   round(effectCI_inf, digits=4), ";", round(effectCI_sup, digits=4), "].", "\n")

if (effect_pvalue < 0.05){
    cat("L'IC autour de l'effet n'inclut pas zéro.")
} else {
    cat("L'IC autour de l'effet inclut zéro.")
}
```

### On peut également visualiser l'IC de l'effet *Vision* sur RMS_x, graphiquement !

```r
#install.packages("ggdist")
library(ggplot2)
library(dplyr)
library(ggdist)
```

```r
# le data.frame dans la ligne ci-dessous avait été créé avant (voir plus haut) !!
VisEffect_xdata

# moyenne de l'effet Vision : on récupère la valeur dans la 3ème ligne et 2ème colonne
VisEffect_xdata_mean <- VisEffect_xdata[3, 2]
# VisEffect_xdata_mean

# erreur-type de l'effet Vision : on récupère la valeur dans la 3ème ligne et 2ème colonne
VisEffect_xdata_se <- VisEffect_xdata[3, 3]
# VisEffect_xdata_se
```

### La borne inférieure de *l'intervalle de confiance à 95% centré sur l'effet* ne recouvre pas zéro

```r
garantie=0.95
alpha_bilat=(1-garantie)/2

# les ddl de l'effet ont été précédemment calculés (voir plus haut) !!
ddl <- effect_ddl

VD_mean <- VisEffect_xdata_mean
VD_se <- VisEffect_xdata_se

# ceci calcule la valeur du t de Student pour la borne inférieure de l'intervalle
x_lower <- qt(alpha_bilat, ddl, lower.tail = TRUE, log.p = FALSE)
x_lower <- round(x_lower, digits = 4)
# ceci calcule la valeur du t de Student pour la borne supérieure de l'intervalle
x_upper <- qt(1-alpha_bilat, ddl, lower.tail = TRUE, log.p = FALSE)
x_upper <- round(x_upper, digits = 4)

#calcul des bornes de l'IC
IC_inf <- VD_mean+x_lower*VD_se
IC_inf <- round(IC_inf, digits = 4)
IC_sup <- VD_mean+x_upper*VD_se
IC_sup <- round(IC_sup, digits = 4)

options(repr.plot.width=12, repr.plot.height=8)

My_Theme = theme(
  axis.title.x = element_text(size = 16),
  axis.text.x = element_text(size = 14),
  axis.text.y = element_text(size = 14),
  axis.title.y = element_text(size = 16),
   legend.title=element_text(size=16), 
    legend.text=element_text(size=14),
)

priors = tribble(
  ~ dist,      ~ args,
    "student_t", list(ddl, VD_mean, VD_se)
) 

priors %>%
  ggplot(aes(y = 0, dist = dist, args = args)) +
# on utilise les valeurs des bornes, calculées précédemment
  stat_dist_halfeye(aes(fill = stat(IC_inf < (x) & IC_sup > (x))), show.legend=FALSE) +
scale_fill_manual(values = c("grey85", "lightblue")) +
geom_text(x=0.01, y= 1, label="H0", size = 8) + geom_text(x=VD_mean, y= 1, label="Distribution de l'effet", size = 8) +
geom_text(x=IC_inf, y= -0.025, label=round(IC_inf, digits=4), size = 8) + 
geom_text(x=IC_sup, y= -0.025, label=round(IC_sup, digits=4), size = 8) + geom_vline(xintercept = 0, lty=8) +
ggtitle(paste("H0 et distribution du t(", ddl, ") centrée sur l'effet de moyenne ", VD_mean, 
              " avec un intervalle de confiance à la garantie ", garantie, sep = ""))  + My_Theme 
```

### On peut vérifier que l'effet *Vision* sur RMS_x est significatif parce que l’intervalle de confiance ne recouvre pas zéro, raison pour laquelle l’effet est significatif 

#### On utilise un test *t* permettant de comparer l'effet observé à la valeur théorique 0 (zéro).

```r
# on utilise la 3ème colonne du data.frame *vis_effect_RMS_x* qui contient les données
# de l'effet Vision (différence entre yeux fermés et yeux ouverts) pour chaque sujet

effet_vision_x <- vis_effect_RMS_x[3]

# on choisit un test t comparant une moyenne observée à la moyenne théorique mu=0
t_effet_vision_x <- t.test(effet_vision_x, mu=0, alternative="two.sided")

effet_tvalue <- t_effet_vision_x$statistic # la valeur t calculée correspondant à l'effet étudié
effet_ddl <- t_effet_vision_x$parameter # degrés de liberté
effet_pvalue <- t_effet_vision_x$p.value #  valeur p du test de Student

effet_CI <-  t_effet_vision_x$conf.int # intervalle de confiance (IC) autour de l'effet
effetCI_inf <- t_effet_vision_x$conf.int[1] # Valeur de la borne inférieure de l'IC 
effetCI_sup <- t_effet_vision_x$conf.int[2] # Valeur de la borne supérieure 

if (effet_pvalue < 0.05){
    cat("L'effet étudié est significatif au seuil alpha = 0.05, ")
} else {
    cat("L'effet étudié est non-significatif au seuil alpha = 0.05, ")
}

cat("t(", effet_ddl, ") = ", round(effet_tvalue, digits=4), ", p = ", +
    round(effet_pvalue, digits=5), ";", "\n")
    
   cat("l'intervalle de confiance (IC) à 95% autour de l'effet est [", +
   round(effetCI_inf, digits=4), ";", round(effetCI_sup, digits=4), "].", "\n")

if (effet_pvalue < 0.05){
    cat("L'IC autour de l'effet n'inclut pas zéro.")
} else {
    cat("L'IC autour de l'effet inclut zéro.")
}
```
